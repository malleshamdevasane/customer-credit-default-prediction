# Final Analysis: Customer Credit Default Prediction

## 1. Introduction
This project aims to build a machine learning pipeline that accurately predicts customer credit default based on historical financial and behavioral data. The problem is modeled as a binary classification task, with the target variable indicating default (1) or non-default (0). The goal is to improve credit decision processes by identifying potential defaulters with high precision and recall. The dataset, consisting of behavioral and demographic customer profiles, offers a rich context for predictive modeling.

## 2. Data Cleaning and Preparation
### üìÅ Datasets:
- `processed_train_data.csv` (70,000 records, 82 features)
- `processed_test_data.csv` (33,000 records, 80 features)

### üßπ Cleaning Steps:
- Missing values handled with median imputation.
- Categorical variables encoded using Label Encoding.
- Numerical features scaled using RobustScaler.
- Final features aligned to 80 columns in both datasets.

## 3. Exploratory Data Analysis (EDA)
Exploratory analysis was carried out to uncover patterns and correlations in the dataset:

- Class Imbalance: The target class ('default') showed an imbalance, with significantly fewer positive cases (defaults) compared to negatives.

- Feature Distributions: Visualizations such as histograms and boxplots helped in understanding skewness and variability.

- Correlation Matrix: A heatmap of correlations revealed weak to moderate associations among features.

- Outlier Detection: Outliers in credit amounts and payment delays were identified and handled through robust scaling.

## 4. Model Selection and Hyperparameter Tuning
Models used:
- Random Forest: Offers feature importance insights and handles imbalanced datasets well.
- XGBoost: A gradient boosting technique known for its high performance on structured data.
- Decision Tree : Simple and interpretable but prone to overfitting.
- Logistic Regression : A baseline linear model that is interpretable and useful for benchmarking.
- 
Each model was trained with 5-fold cross-validation and hyperparameter tuning using GridSearchCV. Parameters like max_depth, n_estimators, learning_rate, and C were fine-tuned.

## 6. Performance Charts


```python
import matplotlib.pyplot as plt
import numpy as np

models = ['Random Forest', 'XGBoost', 'Decision Tree', 'Logistic Regression']
roc_auc = [0.9633, 0.9576, 0.8808, 0.9057]
f1_scores = [0.8049, 0.7906, 0.6855, 0.7134]

x = np.arange(len(models))
width = 0.35
fig, ax = plt.subplots()
bars1 = ax.bar(x - width/2, roc_auc, width, label='ROC AUC')
bars2 = ax.bar(x + width/2, f1_scores, width, label='F1 Score')

ax.set_ylabel('Scores')
ax.set_title('Model Performance Comparison')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend()

for i in range(len(models)):
    ax.text(x[i] - width/2, roc_auc[i] + 0.005, f'{roc_auc[i]:.3f}', ha='center')
    ax.text(x[i] + width/2, f1_scores[i] + 0.005, f'{f1_scores[i]:.3f}', ha='center')

plt.tight_layout()
plt.show()
```

## 7. Insights and Recommendations
- Random Forest emerged as the top performer with the highest ROC-AUC and F1-score.
- XGBoost was a close second and offers potential for ensemble with Random Forest.
- Logistic Regression had high recall but low precision, implying more false positives.
- Decision Tree underperformed, likely due to its simplicity and lack of generalization.
